```julia
using LinearAlgebra, Plots
```

```julia
using TestImages, Images
```

```julia
img = testimage("airplane_F16")#lena_gray_256")
```

```julia
# Converte a matriz para números e escala de cinza
A = float64.(Gray.(img))
```

```julia
# Calcula a decomposição em valor singular da matriz 
U,S,V = svd(A);
```

```julia
# Dimensão da imagem
dim = size(A,1)
```

```julia
# Vamos ver os valores singulares
plot(S,label="valor singular",xlabel="base",ylabel="vs")
```

```julia
# Vamos montar a imagem original usando os N maiores termos de base
N = 20;

# Evita números inválidos
N = max(1,min(N,dim))
```

```julia
# Inicializa a imagem 
B = zeros(dim,dim);
```

```julia
# Loop para reconstruir a imagem 
for i=1:N
    B .= B .+ S[i]*U[:,i]*V[:,i]'
end
```

```julia
# Mostra a imagem
convert(typeof(A),B)
```
